```python
df = df.sort_values(by=['id', 'date'])
df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4',
       'PO4', 'CL']
```

```python
pollutants
```

```python
print(df[pollutants])
```

```python
df_clean = df[pollutants].dropna()
```

```python
sns.set(style="whitegrid")
```

```python
# Plot 1: Boxplot for distribution of each pollutant
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_clean)
plt.title("Distribution of Pollutant Levels")
plt.ylabel("Concentration")
plt.xlabel("Pollutants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Pairplot for pairwise relationships
sns.pairplot(df_clean)
plt.suptitle("Pairwise Relationships Between Pollutants", y=1.02)
plt.show()

# Plot 3: Heatmap for correlation
plt.figure(figsize=(8, 6))
corr = df_clean.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Pollutants")
plt.show()
```
